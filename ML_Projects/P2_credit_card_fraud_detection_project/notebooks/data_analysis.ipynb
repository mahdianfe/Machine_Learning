# این نوت‌بوک شامل تجزیه و تحلیل داده‌ها و تجسم آن‌ها است.

```python
# Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Data
data = pd.read_csv('../data/creditcard.csv')

# بررسی اجمالی داده‌ها
print(data.head())
print(data.info())
print(data.describe())

# تجزیه و تحلیل توزیع ویژگی‌ها
sns.histplot(data['Amount'], kde=True)
plt.title('Distribution of Transaction Amount')
plt.show()

# توزیع کلاس‌ها (Normal vs Fraud)
sns.countplot(x='Class', data=data)
plt.title('Distribution of Classes (Normal vs Fraud)')
plt.show()

# بررسی همبستگی بین ویژگی‌ها
correlation_matrix = data.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.1f', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```
